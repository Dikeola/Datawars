```python
Importing the libraries

# to read the dataframe
import pandas as pd
# For mathematical calculations
import numpy as np
Reading the dataset as a dataframe

# Reading dataframes
exercises_df = pd.read_csv('exercise_dataset.csv')

exercises_df.head()
ID	Exercise	Calories Burn	Dream Weight	Actual Weight	Age	Gender	Duration	Heart Rate	BMI	Weather Conditions	Exercise Intensity
0	1	Exercise 2	286.959851	91.892531	96.301115	45	Male	37	170	29.426275	Rainy	5
1	2	Exercise 7	343.453036	64.165097	61.104668	25	Male	43	142	21.286346	Rainy	5
2	3	Exercise 4	261.223465	70.846224	71.766724	20	Male	20	148	27.899592	Cloudy	4
3	4	Exercise 5	127.183858	79.477008	82.984456	33	Male	39	170	33.729552	Sunny	10
4	5	Exercise 10	416.318374	89.960226	85.643174	29	Female	34	118	23.286113	Cloudy	3
Discovering the dataset

#show the number of cells in the dataframe
print("exercises dataframe size: ", exercises_df.size)
​
#show the number of records (rows) in the dataframe
print("number of exercises: ", len(exercises_df))
exercises dataframe size:  46368
number of exercises:  3864

exercises_df.dtypes
ID                      int64
Exercise               object
Calories Burn         float64
Dream Weight          float64
Actual Weight         float64
Age                     int64
Gender                 object
Duration                int64
Heart Rate              int64
BMI                   float64
Weather Conditions     object
Exercise Intensity      int64
dtype: object

exercises_df.info()
<class 'pandas.core.frame.DataFrame'>
RangeIndex: 3864 entries, 0 to 3863
Data columns (total 12 columns):
 #   Column              Non-Null Count  Dtype
---  ------              --------------  -----
 0   ID                  3864 non-null   int64
 1   Exercise            3864 non-null   object
 2   Calories Burn       3864 non-null   float64
 3   Dream Weight        3864 non-null   float64
 4   Actual Weight       3864 non-null   float64
 5   Age                 3864 non-null   int64
 6   Gender              3864 non-null   object
 7   Duration            3864 non-null   int64
 8   Heart Rate          3864 non-null   int64
 9   BMI                 3864 non-null   float64
 10  Weather Conditions  3864 non-null   object
 11  Exercise Intensity  3864 non-null   int64
dtypes: float64(4), int64(5), object(3)
memory usage: 362.4+ KB
Activities
Let's enjoy transforming exercises data.
1. Classify age into categories

bins = [0,30,60,float('inf')]
labels = ['Young','Adult','Senior']
exercises_df['AgeGroup'] = pd.cut(exercises_df['Age'],bins=bins,labels=labels)
2. Categorize Actual Weight into Weight Categories

bins = [0,60,80,float('inf')]
labels = ['Lightweight','Normal Weight','Overweight']
exercises_df['WeightCategory'] = pd.cut(exercises_df['Actual Weight'],bins=bins,labels=labels)
3. Categorize Exercise Durations into Meaningful Groups

bins = [float('-inf'),25,50,float('inf')]
labels = ['Sedentary','Moderate','Active']
exercises_df['ExerciseLevel'] = pd.cut(exercises_df['Duration'],bins=bins,labels=labels)
4. Generate Gender dummy variables with a colon separator

gender_dummies = pd.get_dummies(exercises_df['Gender'], prefix='Gender',prefix_sep=':')
5. Create Dummy Variables for Weather Conditions Column and Drop First Categorical Level

Weather_dummies = pd.get_dummies(exercises_df['Weather Conditions'],prefix='Weather',prefix_sep=':',drop_first=True)
6. Categorize Exercises by Intensity Level

bins = [0,0.25,0.50,0.75,1]
labels = ['Low Intensity','Medium Intensity','Above-Medium Intensity','High Intensity']
exercises_df['ExerciseIntensityCategory'] = pd.qcut(exercises_df['Exercise Intensity'],4,labels=labels)
7. Categorize Exercises by Calorie Burn Level

calorie_burn_bin_edges = ['Low Calorie Burn','Moderate Calorie Burn','High Calorie Burn']
exercises_df['CalorieBurnCategory'] = pd.qcut(exercises_df['Calories Burn'],3,labels=calorie_burn_bin_edges)
​
8. Categorize exercises into BMI categories with 3 quantiles

bmi_bin_edges = ['Underweight','Normal Weight','Overweight']
exercises_df['BMICategory']=pd.qcut(exercises_df['BMI'],3,labels=bmi_bin_edges)
​
9. Categorize Exercises by Intensity

heart_rate_bin_edges = ['Resting Zone','Fat-Burning Zone','Cardio Zone']
exercises_df['HeartRateZone'] = pd.qcut(exercises_df['Heart Rate'],3,labels=heart_rate_bin_edges)
10. Analyze the relationship between BMI and exercise intensity or fitness metrics and calculate average Exercise Intensity for each BMICategory

bmi_exercise_frequency = exercises_df.groupby('BMICategory')['Exercise Intensity'].agg('mean')
11. Create a grouped bar chart to visualize the relation between CalorieBurnCategory & Exercise

exercises_count_data = exercises_df.groupby(['CalorieBurnCategory','Exercise']).size().unstack()
calories_exercises_count_chart = exercises_count_data.plot(kind='bar',stacked=True, figsize=(10,6))

The End!
```
